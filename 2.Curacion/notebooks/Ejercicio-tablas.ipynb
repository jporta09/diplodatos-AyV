```python
#primero distingo entre 'sex'

mask_sex1 = tabla_1['Sex']==1
#print(mask_sex1)
dataset1 = tabla_1[mask_sex1]

plt.figure(figsize=(8,5))
sns.lineplot(x = dataset1['Agemos'], y = dataset1['P3'])
sns.lineplot(x = dataset1['Agemos'], y = dataset1['P5'])
sns.lineplot(x = dataset1['Agemos'], y = dataset1['P10'])
sns.lineplot(x = dataset1['Agemos'], y = dataset1['P25'])
sns.lineplot(x = dataset1['Agemos'], y = dataset1['P50'])
sns.lineplot(x = dataset1['Agemos'], y = dataset1['P75'])
sns.lineplot(x = dataset1['Agemos'], y = dataset1['P90'])
sns.lineplot(x = dataset1['Agemos'], y = dataset1['P95'])
sns.lineplot(x = dataset1['Agemos'], y = dataset1['P97'])
plt.title('Masculino')

mask_sex2 = tabla_1['Sex']==2
#print(mask_sex2)
dataset2 = tabla_1[mask_sex2]

plt.figure(figsize=(8,5))
sns.lineplot(x = dataset2['Agemos'], y = dataset2['P3'])
sns.lineplot(x = dataset2['Agemos'], y = dataset2['P5'])
sns.lineplot(x = dataset2['Agemos'], y = dataset2['P10'])
sns.lineplot(x = dataset2['Agemos'], y = dataset2['P25'])
sns.lineplot(x = dataset2['Agemos'], y = dataset2['P50'])
sns.lineplot(x = dataset2['Agemos'], y = dataset2['P75'])
sns.lineplot(x = dataset2['Agemos'], y = dataset2['P90'])
sns.lineplot(x = dataset2['Agemos'], y = dataset2['P95'])
sns.lineplot(x = dataset2['Agemos'], y = dataset2['P97'])
plt.title('Femenino')
```
